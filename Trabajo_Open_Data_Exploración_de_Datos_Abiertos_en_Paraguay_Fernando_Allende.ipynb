# HIPOTESIS

Hipotesis: Las mujeres son más propensas a tener accidentes en diciembre entre las 21:00hs y las 23:59hs.

```python
import matplotlib.pyplot as plt

# Extraer los datos relevantes de data1 para el rango horario y mes específicos
rango_horario = '21:00hs. a 23:59hs.'
mes = 'Diciembre'
accidentes_mujeres = data1.loc[11, rango_horario]

# Calcular la proporción de accidentes que involucran a mujeres en el rango horario especificado
total_accidentes = data1.loc[11, 'TOTAL']
proporcion_mujeres_accidentes = accidentes_mujeres / total_accidentes

# Verificar si la hipótesis es aplicable o no
if accidentes_mujeres == 0:
    print("La hipótesis no es aplicable, ya que no hubo accidentes involucrando mujeres en el rango horario y mes especificados.")
else:
    # Generar el gráfico de pastel para visualizar la proporción de accidentes que involucran a hombres y mujeres
    etiquetas = ['Hombres', 'Mujeres']
    tamaños = [1 - proporcion_mujeres_accidentes, proporcion_mujeres_accidentes]
    colores = ['#ff9999', '#66b3ff']
    explotar = (0, 0.1)  # Solo "explotar" el segundo segmento (es decir, Mujeres)

    plt.pie(tamaños, explode=explotar, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')  # La relación de aspecto igual asegura que el gráfico de pastel se dibuje como un círculo
    plt.title(f'Proporción de Accidentes Involucrando a Mujeres en {mes} ({rango_horario})')
    plt.show()

```

```python
import numpy as np

# Datos de ejemplo - data1
data1 = {
    'MES': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'TOTAL': [141, 119, 135, 147, 139],
    '00:00hs. a 02:59hs.': [3, 1, 3, 2, 4],
    '03:00hs. a 05:59hs.': [8, 11, 10, 11, 10],
    '06:00hs. a 08:59hs.': [28, 24, 22, 18, 23],
    '09:00hs. a 11:59hs.': [20, 12, 28, 28, 34],
    '12:00hs. a 14:59hs.': [20, 20, 19, 28, 11],
    '15:00hs. a 17:59hs.': [25, 22, 32, 29, 17],
    '18:00hs. a 20:59hs.': [20, 9, 15, 25, 28],
    '21:00hs. a 23:59hs.': [17, 9, 6, 6, 12]
}

# Datos de ejemplo - data2
data2 = {
    'MES': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'TOTAL - TOTAL': [121, 82, 128, 118, 106],
    'TOTAL - HOMBRES': [80, 56, 89, 87, 78],
    'TOTAL - MUJERES': [41, 26, 39, 31, 28],
    'HERIDOS - TOTAL': [105, 71, 106, 91, 80],
    'HERIDOS - HOMBRES': [66, 47, 72, 64, 57],
    'HERIDOS - MUJERES': [39, 24, 34, 27, 23],
    'MUERTOS - TOTAL': [16, 11, 22, 27, 26],
    'MUERTOS - HOMBRES': [14, 9, 17, 23, 21],
    'MUERTOS - MUJERES': [2, 2, 5, 4, 5]
}

# Calcular desviación estándar de data1
desviacion_estandar_data1 = {}
for columna in data1.keys():
    if columna != 'MES':
        valores = data1[columna]
        desviacion_estandar_data1[columna] = np.std(valores)

# Calcular desviación estándar de data2
desviacion_estandar_data2 = {}
for columna in data2.keys():
    if columna != 'MES':
        valores = data2[columna]
        desviacion_estandar_data2[columna] = np.std(valores)

print("Desviación estándar de data1:")
print(desviacion_estandar_data1)
print("\nDesviación estándar de data2:")
print(desviacion_estandar_data2)
```

Intervalo de Confianza

```python
import scipy.stats as stats

# Datos de data1
data1_mean = 196.25
data1_std = 89.04738906896709
data1_sample_size = 8

# Cálculo del intervalo de confianza para data1
data1_interval = stats.t.interval(0.95, df=data1_sample_size-1, loc=data1_mean, scale=data1_std / (data1_sample_size**0.5))

# Datos de data2
data2_mean = 419.25
data2_std = 301.5226981505704
data2_sample_size = 8

# Cálculo del intervalo de confianza para data2
data2_interval = stats.t.interval(0.95, df=data2_sample_size-1, loc=data2_mean, scale=data2_std / (data2_sample_size**0.5))

# Imprimir los intervalos de confianza
print("Intervalo de confianza para data1 (95%):", data1_interval)
print("Intervalo de confianza para data2 (95%):", data2_interval)
```
